```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
data_city = pd.read_csv(city_data_to_load)
data_ride = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
data = pd.merge(data_ride, data_city, how ='left', on ='city')
# Display the data table for preview
data.head()

```

```python
#Gathering Bubble Plot Data
city_data = data.groupby(['city','type'])
city_avg_fare = city_data['fare'].mean()
city_tot_rides = city_data['ride_id'].count()
city_tot_drivers = city_data['driver_count'].sum()

plot_df = pd.DataFrame({'Average Fare per City':city_avg_fare,
                   'Total Riders per City':city_tot_rides,
                   'Total Drivers per City':city_tot_drivers
                   })
plot_df = plot_df.reset_index()

urban =plot_df.loc[plot_df['type']=='Urban']
urban_x = urban['Total Riders per City']
urban_y = urban['Average Fare per City']
urban_s = urban['Total Drivers per City']

```

## Bubble Plot of Ride Sharing Data

```python

# Obtain the x and y coordinates for each of the three city types
urban =plot_df.loc[plot_df['type']=='Urban']
urban_x = urban['Total Riders per City']
urban_y = urban['Average Fare per City']
urban_s = urban['Total Drivers per City']

suburban =plot_df.loc[plot_df['type']=='Suburban']
suburban_x = suburban['Total Riders per City']
suburban_y = suburban['Average Fare per City']
suburban_s = suburban['Total Drivers per City']

rural =plot_df.loc[plot_df['type']=='Rural']
rural_x = rural['Total Riders per City']
rural_y = rural['Average Fare per City']
rural_s = rural['Total Drivers per City']


# Build the scatter plots for each city types
fig = plt.figure()
ax1=fig.add_subplot(111)
ax2=fig.add_subplot(111)
ax3=fig.add_subplot(111)

urban = ax1.scatter(urban_x,urban_y, s=urban_s, marker="o", facecolors="lightcoral", edgecolors="black", label = "Urban",alpha = 0.6)
suburban = ax2.scatter(suburban_x,suburban_y, s =  suburban_s, marker="o", facecolors="skyblue", edgecolors="black", label = "Suburban",alpha = 0.6)
rural = ax3.scatter(rural_x,rural_y, s =  rural_s, marker="o", facecolors="gold", edgecolors="black", label = "Rural",alpha = 0.6)
# Incorporate the other graph properties
plt.grid()
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.title('Pyber Ride Sharing Data (2016)')
# Create a legend

plt.legend(title = 'City Types',loc = 'best', markerscale = 0.4)

# Incorporate a text label regarding circle size
textstr = 'Note: Circle size correlates with driver count per city'
plt.text(0.92, 0.5, textstr, fontsize=11, transform=plt.gcf().transFigure)
# Save Figure
plt.savefig("../Images/pyber_rideshare_2016.png", bbox_inches = 'tight')
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
type_df = data.groupby(['type'])
total_fare_type = type_df['fare'].sum()

# Build Pie Chart
labels = ['Rural','Suburban','Urban']
colors = ['gold','skyblue','lightcoral']
explode = [0,0,0.1]
plt.pie(total_fare_type, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title('% of Total Fares by City Type')
# Save Figure
plt.savefig("../Images/tot_fares_citytype.png", bbox_inches = 'tight')
```

```python
# Show Figure
type_df.head()
```

## Total Rides by City Type

```python
# Calculate Type Percents
total_rides_type = type_df['ride_id'].count()

# Build Pie Chart
labels = ['Rural','Suburban','Urban']
colors = ['gold','skyblue','lightcoral']
explode = [0,0,0.1]
plt.pie(total_rides_type, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title('% of Total Rides by City Type')
# Save Figure
plt.savefig("../Images/tot_rides_citytype.png", bbox_inches = 'tight')
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Type Percents
total_drivers_type = type_df['driver_count'].sum()

# Build Pie Chart
labels = ['Rural','Suburban','Urban']
colors = ['gold','skyblue','lightcoral']
explode = [0,0,0.1]
plt.pie(total_drivers_type, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title('% of Total Drivers by City Type')
# Save Figure
plt.savefig("../Images/tot_drivers_citytype.png", bbox_inches = 'tight')
```

```python
# Show Figure
plt.show()
```
